# Dynamic Time Warping

**Dynamic time warping (DTW)** is an algorithm used to measure similarity between two sequences.

Assuming there are two sequences $X=\{x_1, x_2, \cdot, x_n\}$, and $Y=\{y_1, y_2, \cdot, y_m\}$.   
The **wraping matrix** $W$ is an $n$-by-$m$ grid, where each entry $W_{i,j}$ denotes an align point for $x_i$ on $X$ and $y_j$ on $Y$.   
The **wraping path** $P$ is a list of align points $\{W_{i1, j1}, W_{i2, j2},\cdot ,W_{il, jl}\}$, which minimize the distance $X$ and $Y$. 
The optimal path (i_k, j_k) can be calculated by:   
$$D_{min}(i_k,j_k)=\min D_{min}(i_{k-1},j_{k-1})+d(i_k,j_k|i_{k-1},j_{k-1}),$$
where $d$ is the **Euclidean distance**, and $D$ is the **accumulated distance**:
$$D=\sum_k d(i_k, j_k).$$
There are 3 different cases for path moves from $W_{i_{k-1},j_{k-1}}$ to $W_{i_{k},j_{k}}$:
* Horizontal moves: $W_{i_{k-1},j_{k-1}}\rightarrow W_{i_{k-1},j_{k}}$;
* Vertical moves: $W_{i_{k-1},j_{k-1}}\rightarrow W_{i_{k},j_{k_1}}$;
* Diagonal moves: $W_{i_{k-1},j_{k-1}}\rightarrow W_{i_{k},j_{k}}$.

The Euclidean distance between two sequences in distance matrix:

```python
def compute_euclidean_distance_matrix(x, y):
    dist = np.zeros((len(y), len(x)))
    for i in range(len(y)):
        for j in range(len(x)):
            dist[i,j] = (x[j]-y[i])**2
    return dist
```

```python
import numpy as np
import matplotlib.pyplot as plt

# the random data
x = np.random.randint(0, 10, size=20)
y = np.random.randint(0, 10, size=10)
dist_m = compute_euclidean_distance_matrix(x,y)

plt.matshow(dist_m, cmap="YlGnBu")

plt.show()
```

```python
def compute_accumulated_cost_matrix(x, y):
    distances = compute_euclidean_distance_matrix(x, y)
    path = []
    
    D = np.zeros((len(y), len(x)))
    D[0,0] = distances[0,0]
    path.append((len(y)-1, len(x)-1))
 
    
    for i in range(1, len(y)):
        D[i, 0] = distances[i, 0] + D[i-1, 0]  
        
    for j in range(1, len(x)):
        D[0, j] = distances[0, j] + D[0, j-1]
        
    for i in range(1, len(y)):
        for j in range(1, len(x)):
            curr = D[i, j]
            D[i, j] = min(
                D[i-1, j],    # insertion
                D[i, j-1],    # deletion
                D[i-1, j-1]   # match
            ) + distances[i, j]
            

    i = len(y)-1
    j = len(x)-1
  
    while i > 0 or j > 0:
        if i == 0:
            path.append((i, j-1))
            j = j-1
            continue
        
        if j == 0:
            path.append((i-1, j))
            i = i-1
            continue
            
        min_d = min(D[i-1, j], D[i, j-1], D[i-1, j-1])
        if D[i-1, j-1] == min_d:
            path.append((i-1, j-1))
            i = i-1
            j = j-1
        elif D[i-1, j] == min_d:
            path.append((i-1, j))
            i = i-1
        else:
            path.append((i, j-1))
            j = j-1
                
            
                
    return D, path
```
